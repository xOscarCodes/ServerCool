```python
from keras.layers import Dense
from keras.models import Sequential
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data = pd.read_csv('data_format_final.csv')

X = data[['temperature']].values
y = data[[' ac_temp','fan_speed']].values
```

```python
# feature scaling

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(1,)))  # modify input_shape
model.add(Dense(64, activation='linear'))
model.add(Dense(64, activation='linear'))
model.add(Dense(2))
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
new_X = X.reshape(-1, 1)
model.fit(new_X, y, epochs=500, batch_size=32)
```

```python
current_temp = 20.5
input_data = scaler.transform([[current_temp]])
predicted_output = model.predict(input_data)

ac_temperature = predicted_output[0][0]
fan_speed = predicted_output[0][1]
print(ac_temperature)
print(round(fan_speed, 0))
```
